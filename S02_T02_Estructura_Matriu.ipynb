```python
columnasFijas = 100
#A1 = np.random.randint(10, size= (5,5))
deformacionImagen = np.random.rand(654, columnasFijas,3)
#deformacionImagen= np.random.randint(10, size = (654, columnasFijas))
print(deformacionImagen.shape)

deformacionImagen= np.random.randint(10, size = (654, columnasFijas,3))
print(deformacionImagen.shape)
```

```python
 
arrayCeros = np.zeros((654,900-columnasFijas,3))
print('Forma arrayCeros: ', arrayCeros.shape)

Error = np.append(deformacionImagen, arrayCeros, axis=1)
print('Forma Error:      ', Error.shape)
```

```python
imagenDeformada = image + Error
plt.imshow(imagenDeformada)
plt.show()
```

```python
arrayUnos = np.ones((654,900-columnasFijas,3))
print('Forma arrayCeros: ', arrayCeros.shape)

Error = np.append(deformacionImagen, arrayUnos, axis=1)
print('Forma Error:      ', Error.shape)
imagenDeformada = image + Error
plt.imshow(imagenDeformada)
plt.show()
```
