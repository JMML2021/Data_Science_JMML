```python
import numpy as np
np.random.seed(0)

def estadistica (matriz):
    forma = matriz.shape
    if forma[0] ==1:
        print('La media es: ', round(matriz.mean(),2))
    else:
        print('L’array té més d’una dimensió')

num = int(input('Cantidad de numeros a crear? '))
dim = int(input("Dimensió de l'array? "))

x = np.random.rand(dim,num)
print(x)

estadistica(x)
```
